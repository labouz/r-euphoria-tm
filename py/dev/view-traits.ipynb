```python
# create a dictionary that contains the film title as the key and all the traits as the value

film_dict = {}
def make_dict(df):
    for film in range(len(df)):
        # get list of traits for a film
        film_traits = df[df['Title'] == df['Title'][film]]['Trait'].tolist()
        # create dictionary entry for film
        film_dict[df.iloc[film]['Title']] = film_traits        
make_dict(films)   
```

```python
# given dictionary of films and their traits, calculate similarity between the film and each of the other films

# film_sims = pd.DataFrame(columns=['film', 'other_film', 'similar_traits', 'film_traits', 'other_traits'])
film_sims = []
def get_sim(dict):
    for film in dict:
        for other_film in dict:
            if film != other_film:
                # bind to dataframe
                similar_traits = len(set(dict[film]).intersection(set(dict[other_film])))
                # percentage of traits that are similar
                p_sim = similar_traits / len(set(dict[film]))
                new = pd.DataFrame(columns=['film', 'other_film', 'similar_traits', 'film_traits', 'other_traits', 'p_sim'],
                                   data= [[film, other_film, similar_traits, dict[film], dict[other_film], p_sim]])
                # append new to film_sims
                film_sims.append(new)
                # concatenate new to film_sims iteratively
                # pd.concat([film_sims, new])
                # print(film, other_film)
```

```python
get_sim(film_dict)
film_sims = pd.concat(film_sims)
```
