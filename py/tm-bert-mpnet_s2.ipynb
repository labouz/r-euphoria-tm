```python
import pandas as pd
# going to try modeling with raw comments and cleaned comments
data_raw = pd.read_pickle('../dat/s2_rue_comments.pkl')
# data_raw = list(data_raw[0])
data_clean = pd.read_pickle('../dat/corpus.pkl')
# data_clean = list(data_clean[0])
```

```python
import numpy as np
# as series
data_clean = pd.Series(data_clean[0]).dropna()
# remove na
docs = list(data_clean)
```

Using `BERTopic` with different embedding models

---
Attemptd XLM-R-BERT-BASE however this model is deprecated according to huggingface documenation: https://huggingface.co/sentence-transformers/xlm-r-bert-base-nli-stsb-mean-tokens

```python
from bertopic import BERTopic
# model = BERTopic(embedding_model="xlm-r-bert-base-nli-stsb-mean-tokens")
model = BERTopic(embedding_model="all-mpnet-base-v2")
```

```python
topics, probabilities = model.fit_transform(docs)
```

```python
model.get_topic_freq().head()
```

```python
model.get_topic(0)
```

```python
# retry with countvectorizer to remove stopwords
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words='english')
model2 = BERTopic(vectorizer_model=vectorizer_model, embedding_model="all-mpnet-base-v2")
# fit
topics2, probabilities2 = model2.fit_transform(docs)
```
